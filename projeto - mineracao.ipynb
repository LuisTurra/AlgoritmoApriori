```python
import pandas as pd
from apyori import apriori
## remove os warnings
import warnings
warnings.filterwarnings("ignore")
## mostrar todos os valores nas coluna e mostrar todas as colunas e linhas
pd.set_option("display.max_colwidth", -1)
pd.set_option('display.max_columns', None) 
pd.set_option('display.max_rows', None)
```

```python
## carrega o dataset
dados=pd.read_csv(r'baseTratada.csv')
dados
```

```python
## informação do dataset
dados.info()
```

```python
## mudando o type do dataset para melhorar o use de memoria
def optimazeMemory():
    dados['experience_level'] = dados['experience_level'].astype('category')
    dados['job_title'] = dados['job_title'].astype('category')
    dados['salary_in_usd'] = dados['salary_in_usd'].astype('category')
    dados['remote_ratio'] = dados['remote_ratio'].astype('category')
    dados['company_location_traduzido'] = dados['company_location_traduzido'].astype('category')
    dados['company_size'] = dados['company_size'].astype('category')
    return dados
optimazeMemory()
```

```python
## informação do dataset
dados.info()
```

```python
## lista vazia
transacoes = []
## adiciona o dataframe para string numa 'lista'
for i in range(0,len(dados)):
    transacoes.append([str(dados.values[i,j]) for j in range(0,len(dados.columns))])
    
## adiciona a função apriori, configura e treina 
#regras = apriori(transacoes, min_support = 0.02, min_confidence = 0.99,min_lift = 1.2,min_lenght = 2)
regras = apriori(transacoes, min_support = 0.02, min_confidence = 0.80,min_lift = 1.2,min_lenght = 2)

## coloca o treinamento em uma lista
resultados = [list(x) for x in regras]
resultadosformatados = []
## transforma em um dicionario
for j in range(0,len(resultados)):
    resultadosformatados.append([list(x) for x in resultados[j][2]])
    
resultadosformatados
```

```python
## função para transformar o dicionario em dataframe
def inspect(results):
    
    lhs         = [tuple(result[2][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in resultados]
    confidences = [result[2][0][2] for result in resultados]
    lifts       = [result[2][0][3] for result in resultados]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(resultados), columns = ['Left Hand Side', 'Right Hand Side', 'Suporte', 'Confiaça', 'Lift'])

resultsinDataFrame
```

```python
## orderna por suporte
dataframe = resultsinDataFrame.sort_values(by=['Lift'],ascending=False).head(15)
dataframe
```
