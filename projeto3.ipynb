```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
pi=pd.DataFrame()
pi=df
pi.drop(['Unnamed: 0','work_year', 'employment_type','salary','salary_currency','company_location'], axis=1, inplace=True)
pi
```

```python

"""
  2 859 -  50 000
 50 000 - 100 000
100 000 - 300 000
300 000 - 600 000


"""

'''
0 - presencial
50 - parcial
100 - distancia
'''
```

```python
for i in range(0,len(pi)):
    if int(pi['salary_in_usd'][i])>=2859 and int(pi['salary_in_usd'][i])<50000:
        pi['salary_in_usd'][i]='2 859 -  50 000'
    elif int(pi['salary_in_usd'][i])>=50000 and int(pi['salary_in_usd'][i])<100000:
        pi['salary_in_usd'][i]='50 000 - 100 000'
    elif int(pi['salary_in_usd'][i])>=100000 and int(pi['salary_in_usd'][i])<300000:
        pi['salary_in_usd'][i]='100 000 - 300 000'
    elif int(pi['salary_in_usd'][i])>=300000 and int(pi['salary_in_usd'][i])<600000:
        pi['salary_in_usd'][i]='300 000 - 600 000'
    
    if pi['remote_ratio'][i]==0:
        pi['remote_ratio'][i]='presencial'
    elif pi['remote_ratio'][i]==50:
        pi['remote_ratio'][i]='parcial'
    elif pi['remote_ratio'][i]==100:
        pi['remote_ratio'][i]='distancia'
   
```

```python
pi
```

```python
dados=pi
transacoes = []
for i in range(0,len(dados)):
    transacoes.append([str(dados.values[i,j]) for j in range(0,len(dados.columns))])
    
from apyori import apriori

regras = apriori(transacoes, min_support = 0.02, min_confidence = 0.6,min_lift = 1.2,min_lenght = 2)

resultados = list(regras)
resultados

resultados2 = [list(x) for x in resultados]
resultadosformatados = []
for j in range(0,len(resultados2)):
    resultadosformatados.append([list(x) for x in resultados2[j][2]])
    
resultadosformatados
```
